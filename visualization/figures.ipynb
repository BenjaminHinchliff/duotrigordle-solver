```python
from matplotlib.ticker import PercentFormatter

counts = df.Guesses.value_counts()
total = counts.sum()
freqs = counts / total

fig = plt.figure(figsize=(11, 5))
(ax1, ax2) = fig.subplots(1, 2)
p = ax1.bar(counts.index, counts)
ax1.bar_label(p, [f"{f * 100:.4}%" for f in freqs])
ax1.set_xticks(sorted(counts.index))
ax1.set_xlabel("Guesses")
ax1.set_ylabel("Frequency")
ax1.set_title("Distribution of Guesses")
fig.savefig("guesses-dist.png")

times = df["Time (ms)"] / 1000.0

n, bins, p = ax2.hist(times, bins=np.arange(math.floor(times.min()), math.ceil(times.max())), weights=np.ones_like(times) * 100 / len(times))
ax2.bar_label(p, [f"{f * 100:.4}%" for f in n / n.sum()])
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("Frequency")
ax2.set_title("Distribution of Solve Times")
fig.savefig("letter-max-dists.png")
plt.show()
```
