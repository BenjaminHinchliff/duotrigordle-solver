```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
counts = df.Guesses.value_counts()
total = counts.sum()
freqs = counts / total

fig, ax = plt.subplots()
p = ax.bar(counts.index, counts)
ax.bar_label(p, [f"{f * 100:.4}%" for f in freqs])
ax.set_xticks(sorted(counts.index))
ax.set_xlabel("Guesses")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of Guesses")
fig.savefig("guesses-dist.png")
plt.show()
```

```python
from matplotlib.ticker import PercentFormatter

times = df["Time (ms)"] / 1000.0

fig, ax = plt.subplots()
n, bins, p = ax.hist(times, bins=np.arange(17, 23), weights=np.ones_like(times) * 100 / len(times))
ax.bar_label(p, [f"{f * 100:.4}%" for f in n / n.sum()])
ax.set_xlabel("Time (s)")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of Solve Times")
fig.savefig("times-dist.png")
plt.show()
```
